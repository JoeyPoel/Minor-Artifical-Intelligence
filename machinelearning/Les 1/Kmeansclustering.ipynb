Clusteren: het groeperen van vergelijkbare objecten of datapunten op basis van bepaalde kenmerken.

Kunnen jullie voorbeelden bedenken waar je dit zou doen?

K-means clustering is een *unsupervised* machine learning algoritme:
- Je gaat op zoek naar een *K*-aantal clustercentra in de dataset. Een clustercentrum representeert het centrum van een cluster.
- *Means* verwijst naar het middelen van de data om de clustercentra te vinden.

K-means gaat op zoek naar clustercentra door als beginpunt willekeurige punten voor clustercentra te kiezen, en vervolgens iteratief berekeningen uit te voeren om de optimale positie van de clustercentra te vinden:
- Kies willekeurige punten voor een *K*-aantal clustercentra.
- Wijs alle datapunten toe aan het dichtsbijzijnde clustercentrum.
- Bereken het nieuwe centrum van elk cluster als het gemiddelde van de toegewezen datapunten.
- Herhaal de vorige 2 stappen tot er geen veranderingen meer zijn in de toewijzingen.


#### Een voorbeeld in python

1. Importeer libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

2. Genereer een voorbeelddataset met 3 clusters

```python
n_samples = 300
n_features = 2
n_clusters = 3
random_state = 42

X, _ = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)
```

3. Visualiseer de dataset

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Voorbeelddataset")
plt.show()
```

4. Pas K-means clustering toe

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
kmeans.fit(X)
```

5. Krijg de coördinaten van de clustercentra

```python
cluster_centers = kmeans.cluster_centers_
print(cluster_centers)
```

6. Voeg de clustercentra toe aan de plot

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200, marker='X', label='clustercentra')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-means Clustering Resultaat")
plt.legend()
plt.show()
```
