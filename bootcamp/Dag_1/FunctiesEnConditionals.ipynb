# Functies en conditionals

Kijk wat de volgende operators doen:

```python
print( 5 == 6 )
print( 5 != 6 )
print( 5 < 6 )
print( 5 > 6 )
print( 5 >= 6 )
print( 5 <= 6 )
```

Kijk wat de "in" operator doet:

```python
print('s' in 'Mississippi') # single chars
print('sip' in 'Mississippi') # substrings
print('mis' in 'Mississippi') # case-sensitive
```

3. Maak een functie die 2 string variabelen aan elkaar plakt en print

```python
def function(string1, string2):
    if isinstance(string1, str) and isinstance(string2, str):
        outcome = string1 + string2
        print(outcome)

function("Hello ", "world")
function("Hello ", 1)

```

4. Maak een functie die 2 integers bij elkaar optelt

```python
def function(int1, int2):
        outcome = int1 + int2
        print(outcome)

function(1, 1)
```

Als integer 2 groter dan integer 1 is

```python
def function(int1, int2):
        if(int1 < int2):
            outcome = int1 + int2
            print(outcome)

function(1, 1)
function(1, 2)
function(2, 1)

```

Als de integers gelijk zijn

```python
def function(int1, int2):
        if(int1 == int2):
            outcome = int1 + int2
            print(outcome)

function(1, 1)
function(1, 2)
```
