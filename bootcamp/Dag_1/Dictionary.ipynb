# Dictionary

Maak simpele dictionary a = { 'a' : 1, 'b' : 20}

```python
# Maak een simpele dictionary
a = {'a': 1, 'b': 20}

# Voeg 'c' = 5 toe
a['c'] = 5

# Controleer of de key 'c' nu in de dictionary zit
if 'c' in a:
    print("De key 'c' zit nu in de dictionary.")
else:
    print("De key 'c' zit niet in de dictionary.")

# Print de keys & values
print("Keys:", list(a.keys()))
print("Values:", list(a.values()))

# Loop over de values
print("Ongesorteerde lijst")
for key in a:
    value = a[key]
    print(f"{key}: {value}")


# Loop over de gesorteerde keys
print("Gesorteerde lijst op basis van keys")
for key in sorted(a.keys()):
    value = a[key]
    print(f"{key}: {value}")

# Loop over de gesorteerde values en print de bijbehorende key
print("Gesorteerde lijst op basis van values")
for value in sorted(a.values()):
    for key, val in a.items():
        if val == value:
            print(f"{key}: {value}")

```

Schrijf functie die een dict gebruikt om aantal letters te tellen in een zin

```python
def count_letters(sentence):
    # Initialiseer een lege dictionary om de lettertellingen bij te houden
    letter_count = {}

    # Loop door elke karakter in de zin
    for char in sentence:
        # Negeer spaties
        if char != ' ':
            # Verhoog de telling voor de huidige letter met 1
            # Als de letter nog niet in de dictionary staat, voeg deze dan toe met een telling van 1
            letter_count[char] = letter_count.get(char, 0) + 1

    return letter_count

# Test de functie
zin = "Dit is een testzin om te controleren hoeveel letters er voorkomen."
letter_telling = count_letters(zin)
print("Letter tellingen in de zin:")
print(letter_telling)
```
